# ELASTIC -NET -REGRESSION ON BOSTON HOUSES

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
boston=pd.read_csv("E:/ppts/machine learning/10-elastic-net-reg/boston_houses.csv")
boston
```

```python
# delete the Price(MEDV) column as it is dependent and take it as y
x = boston.drop('MEDV', axis=1)
x
```

```python
y = boston['MEDV'].values
y
```

```python
# take the names of columns

names = x.columns
names
```

```python
# range of columns
rng = range(len(names))
rng
```

```python
# drawing  line plot between range and coefficients in elastic net formula
from sklearn.linear_model import ElasticNet
```

```python
es = ElasticNet(l1_ratio=0.5)
es_coef = es.fit(x,y).coef_
es_coef
```

```python
import matplotlib.pyplot as plt
plt.plot(rng, es_coef)
plt.xticks(rng, names, rotation=60)
plt.ylabel("Coefficients")
plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=1)
 
```

```python
es.fit(x_train,y_train)
```

```python
y_pred=es.predict(x_test)
y_pred
```
